### Make the plot for metrics results (test set).
This is for checking overfitting and underfitting.

```python
def checkTestTrainResults(NormalizedMeanTrain, StandardizedMeanTrain, BetterFEMeanTrain, NormalizedMeanTest, StandardizedMeanTest, BetterFEMeanTest):
    # Get 1 plot.
    fig, ax = plt.subplots(1, 1, figsize=(6, 4))

    # Label locations
    x = np.arange(len(metrics))
    # The width of the bars
    width = 0.35  

    # Transform the mean values of each model por each version of the dataset into numpy arrays.
    _NormalizedMeanTrain   = np.array(NormalizedMeanTrain)
    _StandardizedMeanTrain = np.array(StandardizedMeanTrain)
    _BetterFEMeanTrain     = np.array(BetterFEMeanTrain)

    # Get the mean of the means for each dataset version by model.
    meanMeanTrain = (_NormalizedMeanTrain + _StandardizedMeanTrain + _BetterFEMeanTrain) / 3

    # Transform the mean values of each model por each version of the dataset into numpy arrays.
    _NormalizedMeanTest   = np.array(NormalizedMeanTest)
    _StandardizedMeanTest = np.array(StandardizedMeanTest)
    _BetterFEMeanTest     = np.array(BetterFEMeanTest)

    # Get the mean of the means for each dataset version by model.
    meanMeanTest = (_NormalizedMeanTest + _StandardizedMeanTest + _BetterFEMeanTest) / 3

    # Plot the bars.
    ax.bar(x - width/2, meanMeanTrain, width, label='Train')
    ax.bar(x + width/2, meanMeanTest,  width, label='Test')

    # Make the plot pretty.
    ax.set_xticks(x)
    ax.set_xticklabels(metrics)
    ax.legend()

print("Test vs training set results for model logistic regression")
checkTestTrainResults(_lrNormalizedMeanTrain, _lrStandardizedMeanTrain, _lrBetterFEMeanTrain, _lrNormalizedMean, _lrStandardizedMean, _lrBetterFEMean)
plt.ylim([0.7, 1])
plt.show()

print()
print("Test vs training set results for model k-nearest neighbrods")
checkTestTrainResults(_knNormalizedMeanTrain, _knStandardizedMeanTrain, _knBetterFEMeanTrain, _knNormalizedMean, _knStandardizedMean, _knBetterFEMean)
plt.ylim([0.7, 1])
plt.show()

print()
print("Test vs training set results for model decision tree")
checkTestTrainResults(_dtNormalizedMeanTrain, _dtStandardizedMeanTrain, _dtBetterFEMeanTrain, _dtNormalizedMean, _dtStandardizedMean, _dtBetterFEMean)
plt.ylim([0.7, 1])
plt.show()

print()
print("Test vs training set results for model neural network")
checkTestTrainResults(_nnNormalizedMeanTrain, _nnStandardizedMeanTrain, _nnBetterFEMeanTrain, _nnNormalizedMean, _nnStandardizedMean, _nnBetterFEMean)
plt.ylim([0.7, 1])
plt.show()
```
