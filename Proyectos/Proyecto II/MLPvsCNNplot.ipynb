```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Metrics names
metrics = ["Accuracy", "Precision", "Recall", "F1"]

# MLP results.
MLPCovid = [0.82, 0.71, 0.83, 0.77]
MLPNorma = [0.63, 0.77, 0.63, 0.70]
MLPNeumo = [0.74, 0.68, 0.74, 0.71]
MLPOpaci = [0.82, 0.79, 0.81, 0.81]

# CNN results.
CNNCovid = [0.93, 0.97, 0.88, 0.92]
CNNNorma = [0.98, 0.91, 0.96, 0.93]
CNNNeumo = [0.93, 1.00, 0.90, 0.94]
CNNOpaci = [0.89, 0.89, 0.88, 0.88]

# Class names
_COVID = "Covid"
_NORMA = "Normal"
_NEUMO = "Viral Pneumonia"
_OPACI = "Lung Opacity"

# Model names
_MLP = " (MLP)"
_CNN = " (CNN)"

# Markers for the scatters values.
_COVIDMarker = "o"
_NORMAMarker = "s"
_NEUMOMarker = "^"
_OPACIMarker = "D"

# Sizes for the scatters values.
_Size = 80

# Colors for scatters values.
_MLPColor = "#A93226"
_CNNColor = "#16A085"
#_DTMeanColor = "#2471A3"
#_NNMeanColor = "#884EA0"
```

```python
# Get 1 plot.
figMean, axMean = plt.subplots(1, 1, figsize=(8, 8))

# Plot the means for each model.
axMean.scatter(metrics, MLPCovid, marker=_COVIDMarker, label=(_COVID+_MLP), c=_MLPColor, s=_Size)
axMean.scatter(metrics, MLPNorma, marker=_NORMAMarker, label=(_NORMA+_MLP), c=_MLPColor, s=_Size)
axMean.scatter(metrics, MLPNeumo, marker=_NEUMOMarker, label=(_NEUMO+_MLP), c=_MLPColor, s=_Size)
axMean.scatter(metrics, MLPOpaci, marker=_OPACIMarker, label=(_OPACI+_MLP), c=_MLPColor, s=_Size)

axMean.scatter(metrics, CNNCovid, marker=_COVIDMarker, label=(_COVID+_CNN), c=_CNNColor, s=_Size)
axMean.scatter(metrics, CNNNorma, marker=_NORMAMarker, label=(_NORMA+_CNN), c=_CNNColor, s=_Size)
axMean.scatter(metrics, CNNNeumo, marker=_NEUMOMarker, label=(_NEUMO+_CNN), c=_CNNColor, s=_Size)
axMean.scatter(metrics, CNNOpaci, marker=_OPACIMarker, label=(_OPACI+_CNN), c=_CNNColor, s=_Size)

# Make the legend.
axMean.legend(bbox_to_anchor=(0., 0.175, 1., 1.), loc="upper center", ncol=2, mode="expand", borderaxespad=0.)

# Plot grids.
axMean.grid(True, which="major", axis="y")

# Make the axis below the scatters.
axMean.set_axisbelow(True)

# Show the plot.
plt.show()
```
