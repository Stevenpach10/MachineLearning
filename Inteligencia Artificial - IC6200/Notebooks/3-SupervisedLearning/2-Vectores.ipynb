# Vector
Un vector tiene una direcci√≥n y una magnitud asociados y pueden ser definidos en un espacio N dimensional

```python
import matplotlib.pyplot as plt
from math import sqrt
import numpy as np
```

Los vectores tienen un punto de origen $A = (a_1, a_2, ... ,a_n)$ y $B = (b_1, b_2, ... ,b_n)$, donde el vector:

$u = (b_1, b_2, ... ,b_n) - (a_1, a_2, ... ,a_n)$

Para este caso en particular tenemos como punto $A= (0,0)$ y como punto $B = (4, 3)$

$u = (4, 3)$

* Nota: La resta se realiza entrada por entrada. 

```python
# Definir las entradas del vector partiendo del hecho que inicia en el origen.
x = 4
y = 3
u = (x, y)

# Crear un gr√°fico de dispersi√≥n para el vector
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Magnitud del vector u')

# Configurar los ejes
plt.xlim(0, 5)
plt.ylim(0, 4)
plt.scatter(x, y, color='red', label='Vector u (4, 3)')


# Agregar etiquetas y t√≠tulo
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Plot de un Vector Bidimensional')

# Agregar una cuadr√≠cula
plt.grid(True)

# Agregar leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()
```

## Distancia Manhattan (L1)

La distancia Manhattan entre dos puntos $A = (x_1, y_1, z_1, \ldots, n_1)$ y $B = (x_2, y_2, z_2, \ldots, n_2)$ en un espacio $n$-dimensional se calcula mediante la f√≥rmula:


$\sum_{i=1}^{n}{|x_{i} - y_{i}|}$

A continuaci√≥n un ejemplo con 2 dimensiones donde $A = (0,0)$ y $B = (4, 3)$

```python
# Definir las entradas del vector
A = (0, 0)
B = (4, 3)
u = (B[0] - A[0], B[1] - A[1])
# Configurar los ejes
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.scatter(A[0], A[1], color='red', label='Punto origen')
plt.scatter(B[0], B[1], color='green', label='Punto destino')
line = [(B[0], A[0]), (A[0], A[1])]
plt.plot(line[0], line[1], linestyle='--', color='blue')
print(line)
line = [(B[0], B[0]), (B[1], A[1])]
print(line)
plt.plot(line[0], line[1], linestyle='--', color='blue')



# Agregar etiquetas y t√≠tulo
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Distancia L1 - Manhattan')

# Agregar una cuadr√≠cula
plt.grid(True)

# Agregar leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()
print(f"Distancia/Magnitud L1 del vector U = {abs(A[0] + B[0]) + abs(A[1] + B[1]):.2f}")
```

# Vector unidad

* Es un vector que tiene una longitud o magnitud 1. 
* Se utiliza generalmente para indicar la direcci√≥n de un vector sin considerar su longitud. 
* B√°sicamente ayuda a simplificar calculos por que son m√°s simples.

Se puede obtener el vector unidad de un vector $u$ dividiendo el vector $u$ entre su norma.

El vector unitario $\hat{\mathbf{u}}$ de un vector $\mathbf{u}$ se calcula como:

$$
\hat{\mathbf{u}} = \frac{\mathbf{u}}{\lVert \mathbf{u} \rVert}
$$


```python
# Definir las entradas del vector
A = (0, 0)
B = (4, 3)
u = (B[0] - A[0], B[1] - A[1])

L2 = sqrt((A[0] + B[0])**2 + (A[1] + B[1])**2)
#Vector unidad
VectorUnidad = np.asarray(u)/L2

print(f"Entradas del vector unidad = {VectorUnidad[0]:.2f} {VectorUnidad[1]:.2f}")

# Crear un gr√°fico de dispersi√≥n para el vector
plt.quiver(A[0], A[1], u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Magnitud del vector u')

plt.quiver(A[0], A[1], VectorUnidad[0], VectorUnidad[1], angles='xy', scale_units='xy', scale=1, color='red', label='Magnitud del vector unidad')

# Configurar los ejes
plt.xlim(-2, 5)
plt.ylim(-2, 5)
plt.scatter(A[0], A[1], color='red', label='A')
plt.scatter(B[0], B[1], color='green', label='B')


# Agregar etiquetas y t√≠tulo
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Plot de un Vector Bidimensional')

# Agregar una cuadr√≠cula
plt.grid(True)

# Agregar leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()

L2 = sqrt((A[0] + VectorUnidad[0])**2 + (A[1] + VectorUnidad[1])**2)
print(f"Distancia/Magnitud L2 del vector Unidad = {L2:.2f}")
```

# Producto punto
Tambi√©n conocido como producto escalar o producto interno, es una operaci√≥n entre dos vectores en el cual su resultado es un escalar.

El producto punto entre dos vectores $\mathbf{u}$ y $\mathbf{v}$ se calcula como:

$$
\mathbf{u} \cdot \mathbf{v} = \sum_{i=1}^{n} u_i \cdot v_i
$$

Donde $u_i$ y $v_i$ son las componentes correspondientes de los vectores $\mathbf{u}$ y $\mathbf{v}$ en la dimensi√≥n $i$, y $n$ es la dimensi√≥n de los vectores.

$$
\mathbf{u} \cdot \mathbf{v} = u_1 \cdot v_1 + u_2 \cdot v_2 + \ldots + u_n \cdot v_n
$$

A continuaci√≥n vamos a calcular el producto punto de los vectores $v$ y $w$ con entradas $(4, 3)$ y $(4, 0)$ respectivamente.

```python
v = np.array([4, 3])
w = np.array([4, 0])

plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Magnitud del vector v')
plt.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='red', label='Magnitud del vector w')

# Configurar los ejes
plt.xlim(-2, 5)
plt.ylim(-2, 5)
plt.scatter(v[0], v[1], color='red', label='A')
plt.scatter(w[0], w[1], color='green', label='B')


# Agregar etiquetas y t√≠tulo
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Agregar una cuadr√≠cula
plt.grid(True)

# Agregar leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()

# Calculo del producto punto

dot = np.sum(v*w)
print(f"Calculo del producto punto de los vectores u y w = {dot:.2f}")
```

## Propiedad del producto punto 

La identidad del coseno para el producto punto entre dos vectores $\mathbf{u}$ y $\mathbf{v}$ en funci√≥n de sus magnitudes y el √°ngulo $\theta$ es:

$$
\mathbf{u} \cdot \mathbf{v} = \lVert \mathbf{u} \rVert \cdot \lVert \mathbf{v} \rVert \cdot \cos(\theta)
$$

Donde $\lVert \mathbf{u} \rVert$ y $\lVert \mathbf{v} \rVert$ son las magnitudes de los vectores $\mathbf{u}$ y $\mathbf{v}$, respectivamente, y $\theta$ es el √°ngulo entre los vectores.


### ¬øCu√°l es el angulo que est√° entre los dos vectores v y w?

```python
def calcularAngulo(v, w):
    '''
    Funci√≥n que calcula el angulo entre dos vectores
    Primero calcula el producto punto de dos vectores
    Luego calcula las magnitudes de cada vectores
    Y aplica la formula de identidad del coseno del producto punto

    Retorna el resultado del angulo en radianes y en grados
    
    '''
    productoPunto = np.sum(v*w)
    magnitudV = np.sqrt(np.sum(v**2))
    magnitudW = np.sqrt(np.sum(w**2))

    cosAngulo = productoPunto/ (magnitudV * magnitudW)
    angulo_radianes = np.arccos(cosAngulo)
    #Convertir el angulo a grados
    angulo_grados = np.degrees(angulo_radianes)
    return (angulo_radianes, angulo_grados)
```

### Comprobar la formula de identidad de coseno del producto punto
El siguiente ejemplo comprueba la formula de identidad de coseno del producto punto, con la formula de producto punto

```python
v = np.array([4, 3])
w = np.array([4, 0])

magnitudV = np.sqrt(np.sum(v**2))
magnitudW = np.sqrt(np.sum(w**2))
print(f"Magnitud de V = {magnitudV:.2F} - Magnitud de W = {magnitudW:.2F}")

productoPunto = np.sum(v*w)
print(f"Producto punto entre los vectores V y W = {productoPunto:.2F}")

angulo_radianes , angulo_grados = calcularAngulo(v, w)

print(f"Angulo en radianes = {angulo_radianes:.2F} angulo en grados = {angulo_grados:.2F}")

cosDelAngulo = np.cos(np.radians(angulo_grados))
equivalenciaFormula = magnitudV * magnitudW * cosDelAngulo

print(f"Producto punto entre los vectores V y W, definici√≥n = {productoPunto:.2F} y equivalencia de formula {equivalenciaFormula:.2F}")

assert productoPunto == equivalenciaFormula
print("!!!")
```

# Vector co-direccional
Dos vectores son co-direccionales si tiene la misma direcci√≥n aunque tengan diferente magnitud. En otras palabras, si existe un escalar K, no igual a cero, tal que uno de los vectores es producto del escalar por el vector.

$$
\mathbf{u} = K \cdot \mathbf{v}

$$

Por ejemplo el vector unidad de un vector.

## ¬øCu√°l es el angulo de dos vectores si son co-direccionales?
A continuaci√≥n se presenta un ejemplo de dos vectores $v = (4, 3)$ y $u = (0.80, 0.60)$
Se procede a calcular el angulo entre ambos vectores

```python
v = np.array([4, 3])
u = np.array([0.80, 0.60])

_, anguloCodireccionales = calcularAngulo(v, u)

print(f"Angulo entre el vector v y su vector unidad = {anguloCodireccionales:.2F}")
```

¬øCu√°nto es $cos(0)$?


$$
\mathbf{u} \cdot \mathbf{v} = \lVert \mathbf{u} \rVert \cdot \lVert \mathbf{v} \rVert \cdot \cos(0)
$$


$$
\mathbf{u} \cdot \mathbf{v} = \lVert \mathbf{u} \rVert \cdot \lVert \mathbf{v} \rVert
$$

### ¬øQue ocurre si calculamos el producto punto del mismo vector $v$?
Se sabe que es co-direccional

$$
\mathbf{u} \cdot \mathbf{u} = \lVert \mathbf{u} \rVert \cdot \lVert \mathbf{u} \rVert
$$

$$
\mathbf{u} \cdot \mathbf{u} = {\lVert \mathbf{u} \rVert}^2
$$

### Podemos expresar la magnitud/norma L2 en terminos de producto punto.
$$
\sqrt{\mathbf{u} \cdot \mathbf{u}} = \sqrt{{\lVert \mathbf{u} \rVert^2}}
$$

$$
\sqrt{\mathbf{u} \cdot \mathbf{u}} = {\lVert \mathbf{u} \rVert}
$$

```python
v = np.array([4, 3])
magnitudV = np.sqrt(np.sum(v**2))

productoPunto = np.sum(v*v)
equivalencia = np.sqrt(productoPunto)

print(f"Magnitud/norma L2, definici√≥n = {magnitudV:.2F} y reescrita como un producto punto {equivalencia:.2F}")
assert magnitudV == equivalencia
print("!!!")
```

### Vectores ortogonales
Dos vectores son ortogonales si los vectores son perpendicuales, es decir tienen un angulo de 90 grados entre ellos.

```python
v = np.array([4, 3])
w = np.array([3, -4])

plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Magnitud del vector v')
plt.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='red', label='Magnitud del vector w')

# Configurar los ejes
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.scatter(v[0], v[1], color='red', label='A')
plt.scatter(w[0], w[1], color='green', label='B')


# Agregar etiquetas y t√≠tulo
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Plot de un Vector Bidimensional')

# Agregar una cuadr√≠cula
plt.grid(True)

# Agregar leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()
_, angulo = calcularAngulo(v, w)
print(f"Angulo entre vectores V y W perpendiculares = {angulo:.2F} ")
assert 90 == angulo
print("!!!")
```
